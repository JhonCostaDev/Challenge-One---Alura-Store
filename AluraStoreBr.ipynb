### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

lojas = ['loja1', 'loja2', 'loja3', 'loja4']
dfs_lojas = [loja1, loja2, loja3, loja4]
```

```python
for loja in dfs_lojas:
  loja.info()
```

# 1. Análise do faturamento


```python
faturamentos = []
faturamentos.append(loja1['Preço'].sum())
faturamentos.append(loja2['Preço'].sum())
faturamentos.append(loja3['Preço'].sum())
faturamentos.append(loja4['Preço'].sum())


pd.Series(faturamentos, index=lojas)
```

```python
import matplotlib.pyplot as plt
cores = ['red', 'green', 'blue', 'orange']
plt.bar(x=lojas, height=faturamentos, color=cores)
plt.title('Faturamento por Loja', fontsize=14, fontweight='bold')
plt.xlabel(f'Faturamento Total: R${sum(faturamentos):,.2f}')
plt.ylabel('Faturamento (R$)')
for i, valor in enumerate(faturamentos):
    plt.text(x=lojas[i], y=valor, s=f'R${valor:,.2f}', ha='center')
```

```python
# Criando o gráfico de pizza
plt.figure(figsize=(6,6))  # Define o tamanho do gráfico
plt.pie(faturamentos, labels=lojas, autopct='%1.1f%%', startangle=90, colors=cores)

# Adicionando título
plt.title('% Do faturamento por loja')

# Exibindo o gráfico
plt.show()
```

# 2. Vendas por Categoria


```python

categorias_vendas = [loja['Categoria do Produto'].value_counts() for loja in dfs_lojas]

df_categorias_vendas = pd.concat(categorias_vendas, axis=1)
df_categorias_vendas.columns = lojas
df_categorias_vendas
```

```python
df_categorias_vendas.plot(kind='bar', figsize=(12, 6), color=cores)

plt.title('Quantidade de Produtos Vendidos por Categoria em Cada Loja')
plt.xlabel('Categoria do Produto')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.legend(title='Lojas')
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
media_avaliacao = [loja['Avaliação da compra'].mean() for loja in dfs_lojas]

plt.bar(x=lojas, height=media_avaliacao, color=cores)
plt.title('Avaliação por Loja', fontsize=14, fontweight='bold')
plt.xlabel('Lojas')
plt.ylabel('Avaliação')
plt.ylim(3.9, 4.1)
for i, avaliacao in enumerate(media_avaliacao):
    plt.text(x=lojas[i], y=avaliacao, s=f'{avaliacao:.2f}', ha='center')
#media_avaliacao
```

```python
plt.barh(lojas, media_avaliacao, color=cores, height=0.5)

# Adiciona rótulos e título
plt.xlabel('Estrelas')
plt.ylabel('Lojas')
plt.title('Avaliação por Loja')
plt.xlim(3.9, 4.1)
# Exibe o gráfico
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
vendas_lojas = [loja['Produto'].value_counts().reset_index() for loja in dfs_lojas]

mais_vendidos = [loja.nlargest(2, 'count') for loja in vendas_lojas]

mais_vendidos_lojas = pd.concat(mais_vendidos, axis=1, keys = lojas, names=['Loja']).rename(columns={'count': 'quantidade'})

mais_vendidos_lojas
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados da tabela

produtos = [mais_vendidos_lojas[(loja, 'Produto')].tolist() for loja in lojas]

quantidades = [mais_vendidos_lojas[(loja, 'quantidade')].tolist() for loja in lojas]

# Ordenando os produtos e quantidades de cada loja
# dados_ordenados = [sorted(zip(qtd, prod), reverse=True) for qtd, prod in zip(quantidades, produtos)]
# produtos_ordenados = [[produto for _, produto in loja] for loja in dados_ordenados]
# quantidades_ordenadas = [[qtd for qtd, _ in loja] for loja in dados_ordenados]

# Configurando o gráfico
fig, ax = plt.subplots(figsize=(10, 5))

for i, loja in enumerate(lojas):
    ax.barh(produtos[i], quantidades[i],color=cores[i] ,label=loja)

ax.set_xlabel("Quantidade Vendida")
ax.set_title("Top 3 Produtos Mais Vendidos por Loja")
ax.legend(title="Lojas")
plt.show()
```

# 5. Frete Médio por Loja


```python
frete_lojas = [loja['Frete'].mean() for loja in dfs_lojas]
plt.bar(x=lojas, height=frete_lojas, color=cores)
plt.title('Preço médio do frete por Loja R$', fontsize=14, fontweight='bold')
plt.xlabel('Lojas')
plt.ylabel('Frete Médio')
plt.ylim(15.9, 40.1)

for i, frete in enumerate(frete_lojas):
    plt.text(x=lojas[i], y=frete, s=f'R$ {frete:.2f}', ha='center')
frete_lojas
```
